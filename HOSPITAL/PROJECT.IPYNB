# patient


I created a Person class to serve as the base class for a Patient class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
#function to view info of person
    def view_info(self):
        return f"Name: {self.name}, Age: {self.age}"
```

```python
class Patient(Person):
#create subclass Patient that inherits from Person
    def __init__(self, name, age, medical_record):
        #call the constructor of the parent class Person
        super().__init__(name, age)
        #initialize medical_record attribute
        self.medical_record = medical_record

    def view_record(self):
       #FUNCTION TO VIEW MEDICAL RECORD OF PATIENT
        return f"Medical Record for {self.name}: {self.medical_record}"
```

```python
class Staff(Person):
    """
    Represents a staff member in the hospital.
    Inherits common attributes from Person.
    """
    def __init__(self, name, age, position):
        # Call the constructor of the parent class (Person)
        super().__init__(name, age)
        # Staff-specific attribute
        self.position = position

    def view_info(self):
        """
        Returns detailed information about the staff member.
        Overrides Person.view_info().
        """
        return f"Staff Name: {self.name}, Age: {self.age}, Position: {self.position}"
```

```python
class Department:
    '''define each department in the hospital and manage the number 
       of patient and staff members in each department'''

    def __init__(self, name):
        self.name = name
        self.patients = []
        self.staff = []
    
    def add_patient(self, patient: Patient):
        '''Function used to manage the patients and add them for certain Department
        
        :param patient: the patient will be added to the department.
        :return None:'''

        self.patients.append(patient)
        print(f"Patient '{patient.name}' added to {self.name} department.")

    def add_staff(self, staff_member: Staff):
        '''Function used to manage the staff members and add them for certain Department
        
        :param staff_member: the member will be added to the department.
        :return None:'''

        self.staff.append(staff_member)
        print(f"Staff '{staff_member.name}' added to {self.name} department.")
```

```python
class Hospital:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.departments = []

    def add_department(self, department: Department):
        self.departments.append(department)
        print(f"Department '{department.name}' added to {self.name}.")


if __name__ == "__main__":

    # create hospital object
    my_hospital = Hospital("City Hospital", "Main St")

    # make the department and assign it to the hospital
    cardiology = Department("Cardiology")
    my_hospital.add_department(cardiology)

    # make patient object and add it to the department
    alice = Patient("Alice", 30, "No known allergies")
    cardiology.add_patient(alice)

    # make Staff member object and add it to the department
    dr_smith = Staff("Dr. Smith", 45, "Cardiologist")
    cardiology.add_staff(dr_smith)
    
    print(f"Patient Record: {alice.medical_record}")
    print(dr_smith.view_info())
```
